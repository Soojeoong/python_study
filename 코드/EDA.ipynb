# 클러스터 분석

```python
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('data/young_survey.csv')
df.head()
```

```python
interests = df.loc[:, 'History': 'Pets']
interests.head()
```

```python
#관심사 상관관계
corr = interests.corr()
corr
```

```python
corr['History'].sort_values(ascending=False)
```

```python
sns.clustermap(corr)
```
