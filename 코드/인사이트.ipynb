# 카테고리로 분류

```python
df = pd.read_csv('data/laptops.csv')
df.head()
```

```python
brand_nation = {
    'Dell' : 'U.S',
    'Apple' : 'U.S',
    'Acer' : 'Taiwan',
    'HP' : 'U.S',
    'Lenovo' : 'China',
    'Alienware' : 'U.S',
    'Microsoft' : 'U.S',
    'Asus' : 'Taiwan'
}
```

```python
df['brand'].map(brand_nation) #Dell -> U.S로 바꿈
```

```python
df['brand_nation'] = df['brand'].map(brand_nation)
```

```python
df
```

```python
nation_groups = df.groupby('brand_nation')
type(nation_groups)
```

```python
nation_groups.first() #첫번째 데이터
```

```python
nation_groups.describe()
```

```python
nation_groups.plot(kind='box', y='price')
```

# 데이터 합치기

- inner join
- left outer join
- right outer join
- full outer join

```python
price_df = pd.read_csv('data/vegetable_price.csv')
quantity_df = pd.read_csv('data/vegetable_quantity.csv')
```

```python
price_df
```

```python
quantity_df
```

```python
pd.merge(price_df, quantity_df, on='Product') #inner join
```

```python
pd.merge(price_df, quantity_df, on='Product', how='left') #left outer join
```

```python
pd.merge(price_df, quantity_df, on='Product', how='right') # right outer join
```

```python
pd.merge(price_df, quantity_df, on='Product', how='outer') # full outer join
```
