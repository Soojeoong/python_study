```python
iphone_df[['디스플레이', 'Face ID']] = 'x'
iphone_df
```

```python
iphone_df.loc[['iPhone 7', 'iPhone X']] = 'o'
iphone_df
```

```python
iphone_df.loc['iPhone 7':'iPhone X'] = 'a'
iphone_df
```

```python
iphone_df.iloc[[1,3], [1,4]] = 'V'  #[행], [열]
iphone_df
```

# DataFrame에 값 추가/삭제

```python
iphone_df = pd.read_csv('data/iphone.csv', index_col=0)
iphone_df
```

## 새로운 row 생성

```python
iphone_df.loc['iPhone XR'] = ['2018-10-26', 6.1, '3GB', 'iOS 12.0.1', 'Yes']
iphone_df
```

## 새로운 컬럼 생성

```python
iphone_df['제조사'] = 'Apple' #새로운 컬럼 생성
iphone_df
```

```python
iphone_df.drop('iPhone XR', axis='index', inplace=False) #index : row 삭제
```

```python
iphone_df #inplace=False : 기존 df에는 영향 X
```

```python
iphone_df.drop('iPhone XR', axis='index', inplace=True)
```

```python
iphone_df #바로 삭제됨
```

```python
iphone_df.drop('제조사', axis='columns', inplace=True)
iphone_df
```

```python
iphone_df.drop(['iPhone 7', 'iPhone 8', 'iPhone X'], axis='index', inplace=True)
```

```python
iphone_df
```

# index/column 설정하기

```python
liverpool_df = pd.read_csv('data/liverpool.csv', index_col=0)
liverpool_df
```

## 컬럼명 바꾸기

```python
liverpool_df.rename(columns={'Position': 'position'})
```

```python
liverpool_df #기존 df에 영향 X
```

```python
liverpool_df.rename(columns={'Position': 'position'}, inplace=True)
```

```python
liverpool_df
```

```python
liverpool_df.rename(columns={'Born':'born', 'Number':'number', 'Nationality':'nationality'}, inplace=True)
```

```python
liverpool_df
```

## index 이름

```python
liverpool_df.index.name = 'Player Name'
```

```python
liverpool_df
```

## row명 바꾸기

```python
liverpool_df.set_index('number')
```

```python
liverpool_df['Player Name'] = liverpool_df.index 
```

```python
liverpool_df
```

```python
liverpool_df.set_index('number', inplace=True)
liverpool_df
```
